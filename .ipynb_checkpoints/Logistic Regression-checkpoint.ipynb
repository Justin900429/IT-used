## Logistic regression

### Using sklearn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn import datasets

# Load the breast dataset
data = datasets.load_breast_cancer()
breast_cancer = pd.DataFrame(data.data, columns=data.feature_names)

# Just use only 2 features for demonstration
breast_cancer = breast_cancer.loc[:, ["mean radius", "mean compactness"]]

display(breast_cancer)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Standarize the data
breast_cancer_std = StandardScaler().fit_transform(breast_cancer)

# Fit the data
model = LogisticRegression().fit(breast_cancer_std, data.target)
# Accuracy of the model
print(f"The accuracy of the model is {model.score(breast_cancer_std, data.target)}")

# Plot the result
plt.figure(figsize=(12, 10))
plt.xlim(-2.5, 4)
plt.ylim(-2, 4)

# Plot the data point
plt.scatter(breast_cancer_std[data.target == 1, 0], 
            breast_cancer_std[data.target == 1, 1], 
            c="red", label="positive")

plt.scatter(breast_cancer_std[data.target == 0, 0], 
            breast_cancer_std[data.target == 0, 1], 
            c="blue", label="negative")

# Plot the seperate plane
x = np.arange(-2, 5)
plt.plot(x, x * model.coef_[0][0] + model.coef_[0][1], 
         label="seperation line")

plt.legend(prop={"size": 15})
```

```python
result = zip(model.predict(breast_cancer_std[20:25]),
          model.predict_proba(breast_cancer_std[20:25])[:, 1])

for count, (class_, prob) in enumerate(result):
    print(f"The correponding probability for label {class_} of {count}th object is {prob}.\n")
```

## Logistic regression - 2

### Make our own logistic regression

```python
import numpy as np

def sigmoid(z):
    '''
    Input:
        z: data(1D array)
        
    Use:
        Count the sigmoid of all point
        
    Return
        Sigmoid of data (1D array)
    '''
    g = np.zeros(z.size)
    g = 1 / (1 + np.exp(-z))

    return g


def compute_cost(theta, X, y):
    '''
    Input:
        theta: parameters of the logistic regression (1D array)
        X: data (2D array)
        y: labels (1D array)
        
    Use:
        Compute the loss of the logistic regression
        
    Output:
        cost: The cost of the prediction (float)      
    '''
    m = y.size
    
    zeta = np.dot(X, theta.T).T
    compute_temp = sigmoid(zeta)
    
    # Count the cost
    cost = (np.dot(-y, np.log(compute_temp).T) - \
            np.dot(np.ones(m) - y, np.log(np.ones(m) - compute_temp).T)) / m
    

    grad = X.T * (compute_temp - y)
    grad = np.mean(grad.T, axis=0)

    return cost


def cout_grad(X, y, iters, alpha=0.01):
    '''
    Input:
        X: data (2D array)
        y: labels (1D array)
        iters: iteration to run (int)
        alpha: learning rate (float, default to 0.01)
        
    Use:
        Get the optimize solution of logistic regression
        
    Output:
        theta: parameters of logistic regression (1D array)
        J_history: Loss function of each history (2D array)  
    '''
    # Get the cost of lost function in each iteration
    J_history = np.zeros(iters)
    
    # Init the theta
    theta = np.zeros(X.shape[1])
    
    # Updata the theta
    for i in range(iters):
        zeta = np.dot(X, theta.T).T
        compute_temp = sigmoid(zeta)
        
        grad = X.T * (y - compute_temp)
        grad = np.mean(grad, axis=1)
        
        theta += alpha * grad
        
        J_history[i] = compute_cost(theta, X, y)
        
    return theta, J_history
```

```python
theta, loss = cout_grad(breast_cancer_std, data.target, 15000, 0.01)
```
