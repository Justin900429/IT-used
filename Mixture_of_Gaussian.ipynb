# Mixture of Gaussian

```python
import numpy as np

# Number of samples to generate
n_samples = 300

# Generate two kinds of Gaussian
# The first one is shifted Gaussian
# The second one is stretched Gaussian

# Generate spherical data centered on (20, 20)
# random.randn will generate normal distrubution data
shifted_gaussian = np.random.randn(n_samples, 2) + np.array([20, 20])

# Generate zero centered stretched Gaussian data
C = np.array([[0., -0.7], [3.5, .7]])
stretched_gaussian = np.dot(np.random.randn(n_samples, 2), C)

# concatenate the two datasets into the final training set
X_train = np.vstack([shifted_gaussian, stretched_gaussian])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Plot the training set
plt.scatter(X_train[:, 0], X_train[:, 1])
```

```python
from sklearn.cluster import KMeans

# Try Kmeans model
model = KMeans(n_clusters=2).fit(X_train)
```

```python
# Plot the Kmeans model
plt.figure(figsize=(12, 6))
labels = model.predict(X_train)
plt.scatter(X_train[:, 0], X_train[:, 1], c=labels)
```

```python
from sklearn import mixture

# Fit a Gaussian Mixture Model with two components
# Here, we make every Gaussian have their own covariance matrix
clf = mixture.GaussianMixture(n_components=2)
clf.fit(X_train)
```

```python
from matplotlib.colors import LogNorm

plt.figure(figsize=(12, 6))

# display predicted scores by the model as a contour plot
x = np.linspace(-20., 30.)
y = np.linspace(-20., 40.)
X, Y = np.meshgrid(x, y)
XX = np.array([X.ravel(), Y.ravel()]).T

# Compute the loglikelihood
Z = -clf.score_samples(XX)
Z = Z.reshape(X.shape)

# Using logspace to get appropriate result
# LogNorm is used to demonstrate colors
CS = plt.contour(X, Y, Z, norm=LogNorm(vmin=1.0, vmax=1000.0),
                 levels=np.logspace(0, 3, 10))


# Set size to 0.8 for better illustration
plt.scatter(X_train[:, 0], X_train[:, 1], s=0.8)

plt.title('Negative log-likelihood predicted by a GMM')
plt.axis('tight')
plt.show()
```
