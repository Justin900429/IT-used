# Decision Tree

## Classification

```python
import pandas as pd
from sklearn.datasets import load_iris

data = load_iris()

# Take only 2 features for illustration
df_data = pd.DataFrame(data.data[:, [0, 3]], columns=[data.feature_names[0], data.feature_names[3]])
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_data.iloc[:, 0], df_data.iloc[:, 1], c=data.target)
```

```python
from sklearn.tree import DecisionTreeClassifier

model_cart = DecisionTreeClassifier(criterion="gini", min_samples_leaf=3).fit(df_data, data.target)
model_id3 = DecisionTreeClassifier(criterion="entropy", min_samples_leaf=3).fit(df_data, data.target)
```

```python
import numpy as np

# Plot the decision boundry
X = df_data.values

# Get the grid value
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))


# Plot the figure

Z = [model_cart, model_id3]
title = ["Using CART", "Using ID3"]

plt.figure(figsize=(12, 6))

for i, z in enumerate(Z):
    temp_z = z.predict(np.c_[xx.ravel(), yy.ravel()])
    temp_z = temp_z.reshape(xx.shape)
    
    plt.subplot(1, 2, i + 1)

    plt.title(title[i])
    # Plot contour with color filled
    cs = plt.contourf(xx, yy, temp_z, cmap=plt.cm.Paired)

    # Plot training data
    plt.scatter(df_data.iloc[:, 0], df_data.iloc[:, 1], c=data.target)
```

```python
from sklearn import tree

# Plot the decision
plt.figure(figsize=(15, 30))

# CART
plt.subplot(2, 1, 1)
plt.title("Decision for CART")
result_cart = tree.plot_tree(model_cart, filled=True)

# ID3
plt.subplot(2, 1, 2)
plt.title("Decision for ID3")
result_id3 = tree.plot_tree(model_id3, filled=True)
```

## Regression

```python
from sklearn.datasets import make_regression

# Generate data
X, y= make_regression(n_features=1, noise=30)


# # 3d plot
# fig = plt.figure()
# ax = plt.axes(projection='3d')
# ax.scatter3D(X[:, 0], X[:, 1], y, c=y)

plt.scatter(X, y)
```

```python
from sklearn.preprocessing import StandardScaler

# Standarize the data
X = StandardScaler().fit_transform(X.reshape(-1, 1))
```

```python
from sklearn.tree import DecisionTreeRegressor

# Fit the regression
model_reg = DecisionTreeRegressor().fit(X, y)
```

```python
# Plot the decision boundry
plt.figure(figsize=(20, 6))
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
xx = np.arange(x_min, x_max, 0.000002)
plt.plot(xx, model_reg.predict(xx.reshape(-1, 1)), label="Decision boundry")


# Plot training data
plt.scatter(X, y, c=y)

plt.legend()
```
